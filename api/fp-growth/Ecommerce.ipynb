```python
import pandas as pd
data_ord = pd.read_csv('../../../instacart_2017_05_01/orders.csv')
data_ord.head()
```

```python
data_ord_pro = pd.read_csv('../../../instacart_2017_05_01/order_products__train.csv')
data_ord_pro.head()
```

```python
data_prod = pd.read_csv('../../../instacart_2017_05_01/products.csv')
data_prod.head()
```

```python
data_pro_prior = pd.read_csv('../../../instacart_2017_05_01/order_products__prior.csv')
data_pro_prior.head()
```

```python
order_details = pd.merge(data_pro_prior, data_ord, on='order_id')
order_details.head()
```

```python
orders_products = pd.merge(order_details, data_prod, on='product_id')
orders_products.head()
```

```python
ord_pro = orders_products.drop(['order_dow', 'days_since_prior_order','aisle_id', 'department_id', 'reordered', 'eval_set', 'user_id'
                     ,'add_to_cart_order'], axis=1)
ord_pro.head()
```

```python
items = {}
for (index, row), i in zip(order_details.iterrows(), range(60000)):
        idx = row['order_id']
        if idx in items.keys():
            items[idx].append(row['product_id'])
        else:
            items[idx] = [row['product_id']]
print(items)
        
```

```python
time_based_items = {}
for (index, row), i in zip(order_details.iterrows(), range(60000)):
        idx_t = row['order_hour_of_day']
        if idx_t not in time_based_items.keys():
            time_based_items[idx_t] = {}
        idx = row['order_id']
        if idx in time_based_items[idx_t].keys():
            time_based_items[idx_t][idx].append(row['product_id'])
        else:
            time_based_items[idx_t][idx] = [row['product_id']]
print(time_based_items)
```

```python
def prepare_data(data):
    prepared_data = []
    for key, val in data.items():
        order = {}
        order['id'] = key
        order['items'] = val
        prepared_data.append(order)
    return prepared_data

def prepare_data_time(data):
    prepared_data = []
    for key, val in data.items():
        order = {}
        order[key] = []
        for k, v in val.items():
            order_ = {}
            order_['id'] = k
            order_['items'] = v
            order[key].append(order_)
        prepared_data.append(order)
    return prepared_data

orders_data = prepare_data(items)
orders_data
time_data = prepare_data_time(time_based_items)
time_data

    
```

```python
import json
with open('orders.json', 'w') as outfile:
    json.dump(orders_data, outfile)
with open('orders_time_based.json', 'w') as outfile:
    json.dump(time_data, outfile)
```
